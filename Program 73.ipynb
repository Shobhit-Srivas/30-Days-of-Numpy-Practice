# Write a script to count the frequency of each character in a string.

```python
class Frequency:
    def __init__(self, text):
        self.text = text

    def count_freq(self):
        seen = {}
        for c in self.text:
            if c not in seen:
                seen[c] = 1
            else:
                seen[c] += 1
        return seen

string = input("Enter a string: ")
obj = Frequency(string)
print(obj.count_freq())
```
