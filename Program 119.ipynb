# Write a script to clean a text file by removing extra spaces and blank lines.

```python
try:    
    file = open("sample.txt", "r")
    lines = file.readlines()
    file.close()
    clean_file = open("cleaned.txt", "w")
    for line in lines:
        cleaned_line = " ".join(line.split())
        if cleaned_line:
            clean_file.write(cleaned_line + "\n")
    clean_file.close()
    print("File cleaned successfully.")
except:
    print("File does not found")
```
