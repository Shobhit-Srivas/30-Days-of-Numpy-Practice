# Write a function to find the missing number from a list of 1 to N.

```python
def find_missing_number(numbers, n):
    total = n * (n + 1) // 2
    actual_sum = sum(numbers)
    return total - actual_sum

nums = list(map(int, input("Enter numbers seperated by space: ").split()))
print("Missing Number:", find_missing_number(nums, 5))
```
