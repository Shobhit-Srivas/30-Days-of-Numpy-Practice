# Write a program that accepts a sentence and calculates the number of upper and lower case letters.

```python
class Sentence:
    def __init__(self, text):
        self.text = text
    def Letter(self):
        lower = upper = 0
        for ch in self.text:
            if ch.islower():
                lower += 1
            elif ch.isupper():
                upper += 1
        print(f"Lower is: {lower}\nUpper is: {upper}")
        
string = input("Enter a string: ")
obj = Sentence(string)
obj.Letter()
```
