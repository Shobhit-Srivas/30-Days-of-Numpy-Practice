# Find the 25th, 50th, and 75th Percentiles Using NumPy

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
percentile25 = np.percentile(arr, 25)
percentile50 = np.percentile(arr, 50)
percentile75 = np.percentile(arr, 75)
print(f"25th Percentile is: {percentile25}")
print(f"50th Percentile is: {percentile50}")
print(f"75th Percentile is: {percentile75}")
```
