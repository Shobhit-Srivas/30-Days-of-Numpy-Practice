# Find maximum and minimum values and their indices.

```python
import numpy as np
arr = np.array([1, 10, 5, 85, 9, 5])
max_value = arr.max()
max_index = arr.argmax()
min_value = arr.min()
min_index = arr.argmin()
print(f"Maximum Value is {max_value} and their Maximum Index is {max_index}")
print(f"Minimum Value is {min_value} and their Minimum Index is {min_index}")
```
