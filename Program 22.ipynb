# Create a program to print all Armstrong numbers between 1 to 1000.

```python
num = int(input("Enter a number: "))
temp = num
power = len(str(num))
sum = 0
while temp > 0:
    digit = temp % 10
    sum += digit ** power
    temp //= 10
if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```
