# Create a nested dictionary to represent student records.

```python
students = {
    101: {
        "name": "Shobhit",
        "age": 21,
        "course": "Python"
    },
    102: {
        "name": "Rahul",
        "age": 22,
        "course": "Data Science"
    },
    103: {
        "name": "Aman",
        "age": 20,
        "course": "Machine Learning"
    }
}
for roll_no, details in students.items():
    print("Roll No:", roll_no)    
    for key, value in details.items():
        print(key, ":", value) 
    print()
```
