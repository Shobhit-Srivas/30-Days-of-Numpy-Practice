# Create a program that removes all duplicate characters from a string.

```python
class RemoveDuplicates:
    def __init__(self, text):
        self.text = text
    def remove_duplicate_characters(self):
        result = ""
        for char in self.text:
            if char not in result:
                result += char
        return result

string = input("Enter a string: ")
obj = RemoveDuplicates(string)
print("String after removing duplicate characters:")
print(obj.remove_duplicate_characters())
```
