# Create a function that removes all HTML tags from a string.

```python
import re
def remove_html_tags(text):
    clean_text = re.sub(r"<.*?>", "", text)
    return clean_text
html_string = "<h1>Welcome</h1><p>Python Programming</p>"
result = remove_html_tags(html_string)
print("Original String :", html_string)
print("Without HTML Tags :", result)
```
