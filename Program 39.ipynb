# Write a function to count the frequency of elements in a list.

```python
def Frequency(arr):
    seen = {}
    for x in arr:
        if x not in seen:
            seen[x] = 1
        else:
            seen[x] += 1
    print(seen)
numbers = list(map(int, input("Enter numbers seperated by space: ").split()))
Frequency(numbers)
```
