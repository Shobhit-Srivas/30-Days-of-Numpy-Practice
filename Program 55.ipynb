# Write a program to find the second highest number in a list.

```python
def Highest(arr):
    highest = second_highest = float('-inf')
    for x in arr:
        if x >  highest:
            second_highest = highest
            highest = x
        elif x > second_highest and x != highest:
            second_highest = x
    print(f"Second Highest Number is {second_highest}")
    
numbers = list(map(int, input("Enter numbers seperated by space: ").split()))
Highest(numbers)
```
