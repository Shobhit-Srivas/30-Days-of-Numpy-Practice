# Write a function to find common elements in two lists.

```python
def Common(arr1, arr2):
    print(f"The Common Element in Two lists are: {set(arr1).intersection(set(arr2))}")
num1 = list(map(int, input("Enter numbers seperated by space: ").split()))
num2 = list(map(int, input("Enter numbers seperated by space: ").split()))
Common(num1, num2)
```
