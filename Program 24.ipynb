# Create a calculator app using if-else.

```python
a, b = list(map(int, input("Enter two numbers seperted by space: ").split()))
opt = input("Enter operator (+, -, *, /)")
if opt == '+':
    print(f"Addition of two number is {a+b}")
elif opt == '-':
    print(f"Subtration of two number is {a-b}")
elif opt == '*':
    print(f"Multiplication of two number is {a*b}")
elif opt == '/':
    print(f"Division of two number is {a/b}")
```
