# Write a script to check if a string is a valid URL.

```python
import re
def validate_url(url):
    pattern = r"^(https?:\/\/)?(www\.)?[a-zA-Z0-9-]+\.[a-zA-Z]{2,}(\S*)?$"
    if re.match(pattern, url):
        return "Valid URL"
    else:
        return "Invalid URL"

url = input("Enter a URL: ")
print(validate_url(url))
```
