# Write a function to generate acronyms from a sentence.

```python
def generate_acronym(sentence):
    words = sentence.split()
    acronym = ""
    for word in words:
        acronym += word[0].upper()
    return acronym

text = input("Enter a string: ")
result = generate_acronym(text)
print("Sentence :", text)
print("Acronym  :", result)
```
