# Write a Python program to read a file and display all lines that contain a given keyword.

```python
try:
    keyword = input("Enter keyword to search: ")
    file = open("sample.txt", "r")
    print("\nLines containing the keyword:\n")
    for line in file:
        if keyword.lower() in line.lower():
            print(line.strip())
    file.close()
except:
    print("File does not found")
```
