# Write a function to remove punctuation from a string.

```python
import string
def remove_punctuation(text):
    result = ""
    for char in text:
        if char not in string.punctuation:
            result += char
    return result

sentence = "Hello! How are you today?"
clean_text = remove_punctuation(sentence)
print("Original String :", sentence)
print("Without Punctuation :", clean_text)
```
