# Find the transpose of a matrix and then multiply it with the original matrix.

```python
import numpy as np
arr = np.arange(1, 10).reshape(3, 3)
transpose = arr.T
print(f"Original Matrix\n: {arr}")
print(f"Transpose Matrix\n: {transpose}")
print(f"Multiply of transpose and original matrix\n: {np.dot(arr, transpose)}")
```
