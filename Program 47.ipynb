# Write a function to count vowels in a string.

```python
def Count(string):
    count = 0
    st = string.lower()
    for x in st:
        if x == 'a' or x == 'e' or x == 'i' or x == 'o' or x == 'u':
            count += 1
    print(f"Total Vowels are {count}")
    
st = input("Enter a string: ")
Count(st)
```
