# Write a script that converts camelCase to snake_case.

```python
def camel_to_snake(text):
    result = ""
    for char in text:
        if char.isupper():
            result += "_" + char.lower()
        else:
            result += char
    return result

string = input("Enter a string: ")
converted = camel_to_snake(string)
print("Camel Case :", string)
print("Snake Case :", converted)
```
