# Handle multiple exceptions in a single try block.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    file = open("sample.txt", "r")
    print(file.read())
    
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
except FileNotFoundError:
    print("File not found")
```
