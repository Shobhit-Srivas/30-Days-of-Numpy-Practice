# Swap the first and last rows of a matrix and also swap the first and last columns of a matrix.

```python
import numpy as np
arr = np.arange(1, 10).reshape(3, 3)
print(f"Original Matrix\n: {arr}")
arr[[0, -1]] = arr[[-1, 0]]
print(f"After Swaping Rows Matrix\n: {arr}")
arr = np.arange(1, 10).reshape(3, 3)
print(f"Original Matrix\n: {arr}")
arr[:, [0, -1]] = arr[:, [-1, 0]]
print(f"After Swaping Columns Matrix\n: {arr}")
```
