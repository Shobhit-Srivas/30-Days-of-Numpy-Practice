# Write a program that accepts a string and counts vowels and consonants.

```python
string = input("Enter a sting: ")
vow = con = 0
string = string.lower()
for ch in string:
    if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u':
        vow += 1
    elif ch == " ":
        pass
    else:
        con += 1
print(f"Total Vowels are {vow} and total consonants are {con}")
```
