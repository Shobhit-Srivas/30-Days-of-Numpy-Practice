# Write a program that converts a sentence to Pig Latin.

```python
def pig_latin(sentence):
    words = sentence.split()
    result = []
    for word in words:
        pig_word = word[1:] + word[0] + "ay"
        result.append(pig_word)
    return " ".join(result)

text = "hello world"
converted = pig_latin(text)
print("Original Sentence :", text)
print("Pig Latin Sentence:", converted)
```
